#### How do we pick our team?

In fantasy baseball, each player can only belong to one team (just like in real life). You will be selecting your team with a partner through an **auction draft**. ## Introduction

Based on the way points are to be assigned in this fantasy league, we opted for a relatively simple way of valuing players. We opted to choose a few statistics for pitchers and batters, calculate where each player falls in percentile for those statistics, then weight those percentiles and sum them to get a "score" for the player. We then scale that score to get a dollar cost.

## Data and Statistics

Data was gathered for the years 2016, 2017, and 2018. Statcast data was used for pitching and FanGraphs data for batting.

For batters we chose OPS, BA, and caught stealing per hit. These columns were then converted to their percentile values. Score is calculated as 2xOPS percentile + 1.6xBA percentile - 0.3xsteals per hit percentile. These weights were roughly decided based on how points are assigned in the fantasy scheme. This score was then scaled by a factor of 11 to get a dollar value.

For pitchers we chose IP per game, HR per batter, strike percentage, and HBP and walks per batter. We chose weights of 3, -1, 1.8, and -0.4 respectively. This score was then scaled by 9 to get a dollar value.

## Limitations

Our way of converting score to monetary value does not take into account the max of \$250 compared with the number of players that can be drafted. This means our conversion is a very rough estimate in this regard and so our monetary values cannot necessarily be trusted to be exact. Our method of scoring players is also imperfect as we are just applying rough weights to percentiles based on what appear to be important statistics according to the league scoring method.

```python
import pandas as pd
from pybaseball import playerid_lookup
from pybaseball import playerid_reverse_lookup
from pybaseball import batting_stats
import numpy as np

df_pitches = pd.read_csv("pitches_2016_2018.csv")
```

```python
def get_percentile_column(series):
    return series.rank(pct=True)

def get_batter_score_column(df,
                            ops_column="ops_percentile",
                            ba_column="ba_percentile",
                            steal_column="steal_percentile",
                            ops_weight=2,
                            ba_weight=1.6,
                            steal_weight=-0.3):
    return ops_weight*df[ops_column] + \
           ba_weight*df[ba_column] + \
           steal_weight*df[steal_column]

def get_pitcher_score_column(df,
                             ip_per_game_column="ip_per_game_percentile",
                             hr_column="hr_percentile",
                             strike_rate_column="strike_rate_percentile",
                             hbp_and_walk_column="hbp_and_walk_percentile",
                             ip_per_game_weight=3,
                             hr_weight=-1,
                             strike_rate_weight=1.8,
                             hbp_and_walk_weight=-0.4):
    return ip_per_game_weight*df[ip_per_game_column] + \
           hr_weight*df[hr_column] + \
           strike_rate_weight*df[strike_rate_column] + \
           hbp_and_walk_weight*df[hbp_and_walk_column]
```

```python
def make_ip_per_game(df):
    numgames_numinnings = df.groupby(["pitcher", "game_date"])["inning"].nunique().to_frame().reset_index().groupby("pitcher").agg({"game_date": "nunique", "inning": "sum"})
    return numgames_numinnings["inning"] / numgames_numinnings["game_date"]
```

```python
def make_hr_per_batter(df):
    home_runs = df.loc[df["events"] == "home_run"].groupby(["pitcher"]).size()
    num_batters = df.loc[df["events"].notna()].groupby(["pitcher"]).size()
    return home_runs / num_batters
```

```python
def make_strike_rate(df):
    strikes = df.loc[df["type"] == "S"].groupby(["pitcher"]).size()
    pitches = df.groupby(["pitcher"]).size()
    return strikes / pitches
```

```python
def make_hbp_walk_per_batter(df):
    walks_and_hbp = df.loc[(df["events"] == "walk") | 
                           (df["events"] == "hit_by_pitch")].groupby(["pitcher"]).size()
    num_batters = df.loc[df["events"].notna()].groupby(["pitcher"]).size()
    return walks_and_hbp / num_batters
```

```python
def get_obp_series(df):
    return (df["H"] + df["BB"] + df["HBP"]) / (df["AB"] + df["BB"] + df["HBP"] + df["SF"])

def get_slg_series(df):
    singles = df["H"] - df["HR"] - df["2B"] - df["3B"]
    return (singles + 2*df["2B"] + 3*df["3B"] + 4*df["HR"]) / df["AB"]
```

```python
def get_ops_series(df):
    return get_obp_series(df) + get_slg_series(df)
```

```python
def get_avg_series(df):
    return (df["H"]) / (df["AB"])
```

```python
def create_pitcher_dfs(series,column_name):
    return pd.DataFrame({"pitcher":series.keys(),column_name:series.values})

def create_batter_dfs(series,column_name):
    return pd.DataFrame({"playerID":series.keys(),column_name:series.values})

def setup_pitcher(df_input):
    #df = df_input.copy()
    #df = pd.DataFrame()
    ips = get_percentile_column(make_ip_per_game(df_input))
    hrs = get_percentile_column(make_hr_per_batter(df_input))
    srr = get_percentile_column(make_strike_rate(df_input))
    hbpwb = get_percentile_column(make_hbp_walk_per_batter(df_input))
    names = playerid_reverse_lookup(ips.index)
    names["key_fangraphs"] = names["key_fangraphs"] * 1.0
    return pd.DataFrame({
        "ip_per_game_percentile": ips,
        "hr_percentile": hrs,
        "strike_rate_percentile": srr,
        "hbp_and_walk_percentile": hbpwb,
    }).merge(names[["name_last", "name_first", "key_mlbam"]], left_index=True, right_on="key_mlbam")

def setup_batter():
    df_batting = batting_stats(2016, 2018)
    df_batting = df_batting.loc[df_batting["AB"] > 200]
    df_batting = df_batting[["Name", "AB", "PA", "H", "1B", "2B", "3B", "HR", "BB", "IBB", "SO", "HBP", "SB", "SF"]].groupby(["Name"]).sum()
    df = df_batting.copy()
    df['ops'] = get_ops_series(df)
    df['ba'] = get_avg_series(df)
    df['ops_percentile'] = get_percentile_column(df['ops'])
    df['ba_percentile'] = get_percentile_column(df['ba'])
    
    # modifying by hits for norm
    df['steal_percentile'] = get_percentile_column(df['SB'] / df['H'])
    return df.dropna()

def scale_pitcher(pitcher_score):
    return pitcher_score * 9

def scale_batter(batter_score):
    return batter_score * 11

def query_pitcher(df,last_name):
    batter_query = pd.DataFrame()
    matching_batters = df.loc[df["name_last"].str.contains(last_name)]
    matching_batters["score"] = get_pitcher_score_column(matching_batters)
    matching_batters["cost"] = matching_batters["score"].apply(scale_batter)
    return matching_batters

def query_batter(df,last_name):
    batter_query = pd.DataFrame()
    matching_batters = df.loc[df.index.str.contains(last_name)]
    matching_batters["score"] = get_batter_score_column(matching_batters)
    matching_batters["cost"] = matching_batters["score"].apply(scale_batter)
    return matching_batters
```

```python
df = setup_pitcher(df_pitches)
```

```python
df["cost"] = get_pitcher_score_column(df).apply(scale_pitcher)
```
